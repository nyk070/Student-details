```python
# Student Marks Calculator

# Get student name
name = input("Enter Student Name: ")

# Get marks for 3 subjects
marks = []
for i in range(1, 4):
    mark = int(input(f"Enter Subject {i} Marks: "))
    marks.append(mark)

# Display entered marks
print(f"\nStudent Name: {name}")
for i, m in enumerate(marks, 1):
    print(f"Subject {i} Marks: {m}")
```

```python
# Calculate statistics
total = sum(marks)
average = total / len(marks)
highest = max(marks)
lowest = min(marks)
```

```python
print(f"\nTotal Marks: {total}")
print(f"Average Marks: {average:.2f}")
print(f"Highest Mark: {highest}")
print(f"Lowest Mark: {lowest}")

# Comparisons
print(f"\nSubject 1 > Subject 2: {marks[0] > marks[1]}")
print(f"Subject 3 == Subject 1: {marks[2] == marks[0]}")

# Check if all marks >= 40
all_passed = all(m >= 40 for m in marks)
print(f"\nAll Marks >= 40: {all_passed}")
```
